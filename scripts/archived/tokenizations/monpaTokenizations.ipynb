```python
# import packages
from monpa import utils
from tqdm import tqdm
import codecs
import monpa
import numpy as np
import os
import pandas as pd
```

```python
# configurations
os.getcwd()
RESULT_SAVING_PATH = "../../results/210408_result/210408_dataset_monpaResult.csv"
```

# Import Data

```python
textElement = ""

for fileIndex, fileElement in enumerate(tqdm(os.listdir("../../data/"))):
    file = codecs.open("../../data/" + fileElement, 'r', encoding='utf8', errors='ignore')
    for textIndex, textLines in enumerate(file):
        textElement += textLines
        
    
print("Text Length: ", len(textElement), " \n\n", textElement[-100:])
```

# Further Train Data

```python
monpa.load_userdict("../dicts/monpa_userDefinedDict.txt")
```

# Shorten Sentence to Fit Monpa Cut/Pseg Function

```python
textList = utils.short_sentence(textElement)
# for item in textList:
#     print(item)
```

# Cut Function

```python
# cutResultList = []

# for textIndex, textElement in enumerate(textList):
#     cutResult = monpa.cut(textElement)
#     cutResultList.append(cutResult)

# print(cutResultList)
```

# Pseg Function

```python
psegResultList = []

for textIndex, textElement in enumerate(tqdm(textList)):
    psegResult = monpa.pseg(textElement)
    psegResultList.append(psegResult)
#     print(psegResult)

print(psegResultList[:5])
```

# Result Export

```python
segmentedElementList = []
labelList = []

for listIndex, elementList in enumerate(tqdm(psegResultList)):
    if len(elementList) > 0:
        for elementIndex, elementTuple in enumerate(elementList):
                segmentedElementList.append(elementTuple[0])
                labelList.append(elementTuple[1])


data_exportPseg = pd.DataFrame({
    "Segmented Element":segmentedElementList,
    "POS":labelList
})


print(data_exportPseg.head())
data_exportPseg.to_csv(RESULT_SAVING_PATH, encoding="utf8", index=None)
```
