```python
import spacy
```

```python
nlp1 = spacy.load(R".\model")
```

```python
doc = nlp1("大連佳林體育用品有限公司，成立日期2014年05月27日，所在地遼寧省大連市金州區先進街道東山路1115號")

token_list = []
dep_list = []
pos_list = []

for token in doc:
    token_list.append(token)
    dep_list.append(token.dep_)
    pos_list.append(token.pos_)
    
print(token_list)
print(dep_list)
print(pos_list)
```
