```python
import spacy
```

```python
nlp1 = spacy.load(R"../../../KnowledgeGraph_materials/models_kg/210506_spacy_model")
```

```python
doc = nlp1("1976年，喬布斯與另一位聯合創始人史蒂夫·沃茲尼亞克一起創立了蘋果公司，發明了第一台蘋果電腦，並負責將包括iPod和iPhone在內的蘋果產品成功推向市場")

token_list = []
dep_list = []
pos_list = []

for token in doc:
    token_list.append(token)
    dep_list.append(token.dep_)
    pos_list.append(token.pos_)
    
print(token_list)
print(dep_list)
print(pos_list)
```
