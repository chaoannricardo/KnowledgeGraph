```python
doc = custom_nlp("宇文娥英的母親是隋文帝之女樂平公主楊麗華，在隋文帝楊堅篡北周之前，樂平公主嫁給了北周宣帝宇文贇，並誕下了北周公主宇文娥英")

spacy.displacy.render(doc, style="ent", jupyter=True)

token_list = []
dep_list = []
pos_list = []

for token in doc:
    token_list.append(token)
    dep_list.append(token.dep_)
    pos_list.append(token.pos_)
    
print(token_list)
print(dep_list)
print(pos_list)
```
