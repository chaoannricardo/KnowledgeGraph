```python
doc = custom_nlp("死亡幻覺，是一部頗爲另類的科幻電影，整部影片充滿着美國20世紀80年代YA電影裡的招牌符碼，導演理查德·凱利卻放棄了宣揚美國夢的樂觀美好，改以憂傷的語調來哀悼青春的消逝，附帶超現實般的道德寓言，仿佛就是21世紀的大衛·林奇，敏銳而詩意地在懷舊類型的影片裡注入了新的觀點")

spacy.displacy.render(doc, style="ent", jupyter=True)

token_list = []
dep_list = []
pos_list = []

for token in doc:
    token_list.append(token)
    dep_list.append(token.dep_)
    pos_list.append(token.pos_)
    
print(token_list)
print(dep_list)
print(pos_list)
```
