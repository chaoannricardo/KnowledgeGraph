```python
import spacy
```

```python
nlp1 = spacy.load(R".\output\model-last")
```

```python
doc = nlp1("古往今來，能飾演古龍小說人物「楚留香」的，無一不是娛樂圈公認的美男子，2011年，36歲的張智堯在《楚留香新傳》裡飾演楚留香，依舊帥得讓人無法自拔")

token_list = []
dep_list = []
pos_list = []

for token in doc:
    token_list.append(token)
    dep_list.append(token.dep_)
    pos_list.append(token.pos_)
    
print(token_list)
print(dep_list)
print(pos_list)
```
