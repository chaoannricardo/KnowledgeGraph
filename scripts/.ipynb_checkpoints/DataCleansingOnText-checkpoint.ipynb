```python
import re

sentences_splitter = ["，", "。", "！", "!", "？",  "；", ";", "："]

pattern = "|".join(sentences_splitter)

text = '更進一步則需要寫作或其他可記錄數字的系統，如符木或於印加帝國內用來儲存數據的奇普。歷史上曾有過許多不同的記數系統 \n'

re.split(pattern, text)
```
