```python
# import packages needed
from spacy.lang.zh.examples import sentences 
from tqdm import tqdm
import codecs
import spacy
import numpy as np
import pandas as pd
```

# Import Data

```python
textElement = ""

file = codecs.open("./data/第二十三章半導體製造概論.txt", 'r', encoding='utf8', errors='ignore')

for textIndex, textLines in enumerate(file):
#     textList.append(textLines)
    textElement += textLines
    
```

# Pseg Function

```python
nlp = spacy.load("zh_core_web_trf")

psegResultList = []
allWords = []
allFlags = []
allDep = []

doc = nlp(textElement)
for token in doc:
#         print(token.text, token.pos_, token.dep_)
    allWords.append(token.text)
    allFlags.append(token.pos_)
    allDep.append(token.dep_)
    
```

```python
data_exportPseg = pd.DataFrame({
    "Segmented Element":allWords,
    "POS":allFlags,
    "Dependecies":allDep
})


print(data_exportPseg.head())
data_exportPseg.to_csv("./result/第二十三章半導體製造概論_spacyResult.csv", encoding="utf8", index=None)
```
