```python
# import packages needed
from monpa import utils
from spacy.lang.zh.examples import sentences 
from tqdm import tqdm
import codecs
import spacy
import numpy as np
import os
import pandas as pd
```

# Import Data

```python
textElement = ""

for fileIndex, fileElement in enumerate(tqdm(os.listdir("../../data/"))):
    file = codecs.open("../../data/" + fileElement, 'r', encoding='utf8', errors='ignore')
    for textIndex, textLines in enumerate(file):
        textElement += textLines
        
    
print("Text Length: ", len(textElement), " \n\n", textElement[-100:])
```

# Cuting to small pieces by MONPA

```python
textList = utils.short_sentence(textElement)
```

# Pseg Function

```python
nlp = spacy.load("zh_core_web_trf")

psegResultList = []
allWords = []
allFlags = []
allDep = []

for textIndex, text in enumerate(textList):
    doc = nlp(text)
    for token in doc:
        allWords.append(token.text)
        allFlags.append(token.pos_)
        allDep.append(token.dep_)
    #    print(token.text, token.pos_, token.dep_)
```

```python
data_exportPseg = pd.DataFrame({
    "Segmented Element":allWords,
    "POS":allFlags,
    "Dependecies":allDep
})


print(data_exportPseg.head())
data_exportPseg.to_csv("../../results/210330_result/210330_dataset_spaCyResult.csv", encoding="utf8", index=None)
```
