```python
# import packages
from tqdm import tqdm
import codecs
import jieba
import jieba.posseg as pseg
import numpy as np
import pandas as pd
```

# Import Data

```python
textElement = ""

file = codecs.open("./data/textDoc_0.txt", 'r', encoding='utf8', errors='ignore')

for textIndex, textLines in enumerate(file):
    textElement += textLines
```

# Pseg Function

```python
psegResultList = []

psegResult = pseg.cut(textElement)
psegResultList.append(psegResult)
    

allFlags = []
allWords = []

for pairIndex, pairElement in enumerate(psegResultList):
    for elementIndex, element in enumerate(pairElement):
        try:
            allWords.append(str(element).split("/")[0])
        except:
            allWords.append("")
        
        try:
            allFlags.append(str(element).split("/")[1])
        except:
            allFlags.append("")
        
print(len(allWords))
print(len(allFlags))
```

```python
data_exportPseg = pd.DataFrame({
    "Segmented Element":allWords,
    "Label":allFlags
})


print(data_exportPseg.head())
data_exportPseg.to_csv("./result/jiebaResult_0.csv", encoding="utf8", index=None)
```
